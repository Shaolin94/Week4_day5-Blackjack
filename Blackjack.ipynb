```python
import random

def create_deck():
    deck = []
    suits = ['Hearts', 'Diamonds', 'Clubs', 'Spades']
    values = ['2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King', 'Ace']
    for suit in suits:
        for value in values:
            deck.append((value, suit))
    return deck

def deal_card(deck):
    return deck.pop(random.randint(0, len(deck)-1))

def calculate_hand_value(hand):
    value = 0
    num_aces = 0
    for card in hand:
        if card[0] in ['Jack', 'Queen', 'King']:
            value += 10
        elif card[0] == 'Ace':
            value += 11
            num_aces += 1
        else:
            value += int(card[0])
    while value > 21 and num_aces:
        value -= 10
        num_aces -= 1
    return value

def print_hand(hand):
    for card in hand:
        print(f"{card[0]} of {card[1]}")

def blackjack():
    bet_money = 100
    while True:
        print(f"\nBet Money: ${bet_money}")
        bet = int(input("Place your bet (minimum bet is 1 dollar , or enter 0 to quit): "))
        if bet == 0:
            print("Thanks for spending your money!")
            break
        if bet > bet_money:
            print("You're Broke. Place a lower bet.")
            continue
        
        deck = create_deck()
        random.shuffle(deck)
        player_hand = [deal_card(deck), deal_card(deck)]
        dealer_hand = [deal_card(deck)]
        
        print("\nYour hand:")
        print_hand(player_hand)
        print("\nDealer's hand:")
        print(f"{dealer_hand[0][0]} of {dealer_hand[0][1]}")
        
        player_score = calculate_hand_value(player_hand)
        dealer_score = calculate_hand_value(dealer_hand)
        
        while player_score < 21:
            action = input("\nDo you want to hit or stand? (h/s): ")
            if action == 'h':
                player_hand.append(deal_card(deck))
                print("\nYour hand:")
                print_hand(player_hand)
                player_score = calculate_hand_value(player_hand)
                print(f"\nYour score: {player_score}")
            elif action == 's':
                break
        
        if player_score > 21:
            print("\nYou busted! Dealer wins.")
            bet_money -= bet
            continue
        
        print("\nDealer's hand:")
        print_hand(dealer_hand)
        while dealer_score < 17:
            dealer_hand.append(deal_card(deck))
            dealer_score = calculate_hand_value(dealer_hand)
        
        print(f"\nDealer's score: {dealer_score}")
        
        if dealer_score > 21 or dealer_score < player_score:
            print("\nYou Won The Hand.")
            bet_money += bet
        elif dealer_score > player_score:
            print("\nThe House Won!")
            bet_money -= bet
        else:
            print("\nIts a tie! Push")

blackjack()
```
